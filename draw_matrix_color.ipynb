```python
!gdown 1ZXE0Kdoeecc-qZrNLWqMWwwStRlv5SZn
import pandas as pd
df = pd.read_csv("LBLS_seqs_input_data.csv")
df = df.fillna('')
```

```python
def br_strategy_calculate(df, strategy_class):
  print('bookroll class '+ str(strategy_class))
  df_tmp = df[df['class'] == strategy_class]
  df_tmp = df_tmp.drop(columns=['class'])
  df_tmp = df_tmp.reset_index(drop=True)
  df_tmp['Reading'] = 0
  df_tmp['Highlight'] = 0
  df_tmp['Notetaking'] = 0

  from collections import Counter
  # seuqnce
  for index, row in df_tmp.iterrows():
    values = Counter(row[2:])
    df_tmp.at[index, 'Reading'] = values['Reading']
    df_tmp.at[index, 'Highlight'] = values['Highligt']
    df_tmp.at[index, 'Notetaking'] = values['Notetaking']

  df_tmp = df_tmp[['username', 'weeks', 'Reading', 'Highlight', 'Notetaking']]


  # learning strategy(strategy)0
  for user in set(df_tmp['username']):
    for w in range(1,17):
      # learning strategy0
      if w not in set(df_tmp[df_tmp['username'] == user]['weeks']):
        # 
        data = pd.DataFrame([[user, w, 0, 0, 0]], columns=["username", "weeks", "Reading", "Highlight", "Notetaking"])
        df_tmp = pd.concat([df_tmp, data], ignore_index=True)
  df_tmp = df_tmp.sort_values(by=['username', 'weeks'])
  df_tmp = df_tmp.reset_index(drop=True)

  # bookroll0 readinghighlightnotetaking initialize
  reading = pd.DataFrame(columns=["username", "weeks1", "weeks2", "weeks3", "weeks4", "weeks5", "weeks6", "weeks7", "weeks8",
                      "weeks9", "weeks10", "weeks11", "weeks12", "weeks13", "weeks14", "weeks15", "weeks16"])
  highlight = pd.DataFrame(columns=["username", "weeks1", "weeks2", "weeks3", "weeks4", "weeks5", "weeks6", "weeks7", "weeks8",
                      "weeks9", "weeks10", "weeks11", "weeks12", "weeks13", "weeks14", "weeks15", "weeks16"])
  notetaking = pd.DataFrame(columns=["username", "weeks1", "weeks2", "weeks3", "weeks4", "weeks5", "weeks6", "weeks7", "weeks8",
                      "weeks9", "weeks10", "weeks11", "weeks12", "weeks13", "weeks14", "weeks15", "weeks16"])

  for user in set(df_tmp['username']):
    # data
    reading_data = []
    highlight_data = []
    notetaking_data = []
    reading_data.append(str(user))
    highlight_data.append(str(user))
    notetaking_data.append(str(user))
    for w in range(1,17):
      tmp = df_tmp[(df_tmp['username'] == user) & (df_tmp['weeks'] == w)]
      reading_data.append(tmp['Reading'].iloc[0])
      highlight_data.append(tmp['Highlight'].iloc[0])
      notetaking_data.append(tmp['Notetaking'].iloc[0])

    # datadataframe
    reading = pd.concat([reading, pd.DataFrame([reading_data], columns=["username", "weeks1", "weeks2", "weeks3", "weeks4", "weeks5", "weeks6", "weeks7", "weeks8",
                      "weeks9", "weeks10", "weeks11", "weeks12", "weeks13", "weeks14", "weeks15", "weeks16"]) ])
    highlight = pd.concat([highlight, pd.DataFrame([highlight_data], columns=["username", "weeks1", "weeks2", "weeks3", "weeks4", "weeks5", "weeks6", "weeks7", "weeks8",
                      "weeks9", "weeks10", "weeks11", "weeks12", "weeks13", "weeks14", "weeks15", "weeks16"]) ])
    notetaking = pd.concat([notetaking, pd.DataFrame([notetaking_data], columns=["username", "weeks1", "weeks2", "weeks3", "weeks4", "weeks5", "weeks6", "weeks7", "weeks8",
                      "weeks9", "weeks10", "weeks11", "weeks12", "weeks13", "weeks14", "weeks15", "weeks16"]) ])

  # reading
  from sklearn.cluster import KMeans
  from collections import Counter
  kmeans = KMeans(n_clusters=5).fit(reading.iloc[:,1:])
  centroids = kmeans.cluster_centers_
  predict = kmeans.predict(reading.iloc[:,1:])

  reading['cluster'] = predict
  # reading1.iloc[:,1:].head()
  # mean0nanmean
  # reading = reading.replace(0, np.nan)
  group_count = reading['cluster'].value_counts()
  reading = reading.groupby(['cluster']).mean()
  reading['count'] = group_count.sort_index()


  # highlight
  from sklearn.cluster import KMeans
  from collections import Counter
  kmeans = KMeans(n_clusters=5).fit(highlight.iloc[:,1:])
  centroids = kmeans.cluster_centers_
  predict = kmeans.predict(highlight.iloc[:,1:])

  highlight['cluster'] = predict
  # reading1.iloc[:,1:].head()
  # mean0nanmean
  # highlight = highlight.replace(0, np.nan)
  group_count = highlight['cluster'].value_counts()
  highlight = highlight.groupby(['cluster']).mean()
  highlight['count'] = group_count.sort_index()



  # notetaking
  from sklearn.cluster import KMeans
  from collections import Counter
  kmeans = KMeans(n_clusters=5).fit(notetaking.iloc[:,1:])
  centroids = kmeans.cluster_centers_
  predict = kmeans.predict(notetaking.iloc[:,1:])
  notetaking['cluster'] = predict
  # reading1.iloc[:,1:].head()
  # mean0nanmean
  # notetaking = notetaking.replace(0, np.nan)
  group_count = notetaking['cluster'].value_counts()
  notetaking = notetaking.groupby(['cluster']).mean()
  notetaking['count'] = group_count.sort_index()

  return reading, highlight, notetaking
```

```python
def vs_strategy_calculate(df, strategy_class):
  print('viscode class '+ str(strategy_class))
  df_tmp = df[df['class'] == strategy_class]
  df_tmp = df_tmp.drop(columns=['class'])
  df_tmp = df_tmp.reset_index(drop=True)
  df_tmp['notebook_open'] = 0
  df_tmp['code_speed'] = 0
  df_tmp['copy_paste'] = 0
  df_tmp['Execution'] = 0
  from collections import Counter
  # seuqnce
  for index, row in df_tmp.iterrows():
    values = Counter(row[2:])
    df_tmp.at[index, 'notebook_open'] = values['notebook_open']
    df_tmp.at[index, 'code_speed'] = values['code_speed']
    df_tmp.at[index, 'copy_paste'] = values['copy_paste']
    df_tmp.at[index, 'Execution'] = values['Execution']

  df_tmp = df_tmp[['username', 'weeks', 'notebook_open', 'code_speed', 'copy_paste', 'Execution']]

  # learning strategy(strategy)0
  for user in set(df_tmp['username']):
    for w in range(1,17):
      # learning strategy0
      if w not in set(df_tmp[df_tmp['username'] == user]['weeks']):
        # 
        data = pd.DataFrame([[user, w, 0, 0, 0, 0]], columns=['username', 'weeks', 'notebook_open', 'code_speed', 'copy_paste', 'Execution'])
        # data = pd.DataFrame(columns=['username', 'weeks', 'notebook_open', 'code_speed', 'copy_paste', 'Execution'])
        df_tmp = pd.concat([df_tmp, data], ignore_index=True)
  df_tmp = df_tmp.sort_values(by=['username', 'weeks'])
  df_tmp = df_tmp.reset_index(drop=True)

  # bookroll0 readinghighlightnotetaking initialize
  notebook_open = pd.DataFrame(columns=["username", "weeks1", "weeks2", "weeks3", "weeks4", "weeks5", "weeks6", "weeks7", "weeks8",
                      "weeks9", "weeks10", "weeks11", "weeks12", "weeks13", "weeks14", "weeks15", "weeks16"])
  code_speed = pd.DataFrame(columns=["username", "weeks1", "weeks2", "weeks3", "weeks4", "weeks5", "weeks6", "weeks7", "weeks8",
                      "weeks9", "weeks10", "weeks11", "weeks12", "weeks13", "weeks14", "weeks15", "weeks16"])
  copy_paste = pd.DataFrame(columns=["username", "weeks1", "weeks2", "weeks3", "weeks4", "weeks5", "weeks6", "weeks7", "weeks8",
                      "weeks9", "weeks10", "weeks11", "weeks12", "weeks13", "weeks14", "weeks15", "weeks16"])
  Execution = pd.DataFrame(columns=["username", "weeks1", "weeks2", "weeks3", "weeks4", "weeks5", "weeks6", "weeks7", "weeks8",
                      "weeks9", "weeks10", "weeks11", "weeks12", "weeks13", "weeks14", "weeks15", "weeks16"])

  for user in set(df_tmp['username']):
    # data
    notebook_open_data = []
    code_speed_data = []
    copy_paste_data = []
    Execution_data = []
    notebook_open_data.append(str(user))
    code_speed_data.append(str(user))
    copy_paste_data.append(str(user))
    Execution_data.append(str(user))

    for w in range(1,17):
      tmp = df_tmp[(df_tmp['username'] == user) & (df_tmp['weeks'] == w)]
      notebook_open_data.append(tmp['notebook_open'].iloc[0])
      code_speed_data.append(tmp['code_speed'].iloc[0])
      copy_paste_data.append(tmp['copy_paste'].iloc[0])
      Execution_data.append(tmp['Execution'].iloc[0])
    # datadataframe
    notebook_open = pd.concat([notebook_open, pd.DataFrame([notebook_open_data], columns=["username", "weeks1", "weeks2", "weeks3", "weeks4", "weeks5", "weeks6", "weeks7", "weeks8",
                      "weeks9", "weeks10", "weeks11", "weeks12", "weeks13", "weeks14", "weeks15", "weeks16"]) ])
    code_speed = pd.concat([code_speed, pd.DataFrame([code_speed_data], columns=["username", "weeks1", "weeks2", "weeks3", "weeks4", "weeks5", "weeks6", "weeks7", "weeks8",
                      "weeks9", "weeks10", "weeks11", "weeks12", "weeks13", "weeks14", "weeks15", "weeks16"]) ])
    copy_paste = pd.concat([copy_paste, pd.DataFrame([copy_paste_data], columns=["username", "weeks1", "weeks2", "weeks3", "weeks4", "weeks5", "weeks6", "weeks7", "weeks8",
                      "weeks9", "weeks10", "weeks11", "weeks12", "weeks13", "weeks14", "weeks15", "weeks16"]) ])
    Execution = pd.concat([Execution, pd.DataFrame([Execution_data], columns=["username", "weeks1", "weeks2", "weeks3", "weeks4", "weeks5", "weeks6", "weeks7", "weeks8",
                      "weeks9", "weeks10", "weeks11", "weeks12", "weeks13", "weeks14", "weeks15", "weeks16"]) ])


  # notebook_open
  from sklearn.cluster import KMeans
  from collections import Counter
  kmeans = KMeans(n_clusters=5).fit(notebook_open.iloc[:,1:])
  centroids = kmeans.cluster_centers_
  predict = kmeans.predict(notebook_open.iloc[:,1:])

  notebook_open['cluster'] = predict
  # notebook_open.iloc[:,1:].head()
  # mean0nanmean
  # reading = notebook_open.replace(0, np.nan)
  group_count = notebook_open['cluster'].value_counts()
  notebook_open = notebook_open.groupby(['cluster']).mean()
  notebook_open['count'] = group_count.sort_index()


  # code_speed
  from sklearn.cluster import KMeans
  from collections import Counter
  kmeans = KMeans(n_clusters=5).fit(code_speed.iloc[:,1:])
  centroids = kmeans.cluster_centers_
  predict = kmeans.predict(code_speed.iloc[:,1:])

  code_speed['cluster'] = predict
  # reading1.iloc[:,1:].head()
  # mean0nanmean
  # highlight = highlight.replace(0, np.nan)
  group_count = code_speed['cluster'].value_counts()
  code_speed = code_speed.groupby(['cluster']).mean()
  code_speed['count'] = group_count.sort_index()



  # copy_paste
  from sklearn.cluster import KMeans
  from collections import Counter
  kmeans = KMeans(n_clusters=5).fit(copy_paste.iloc[:,1:])
  centroids = kmeans.cluster_centers_
  predict = kmeans.predict(copy_paste.iloc[:,1:])

  copy_paste['cluster'] = predict
  # reading1.iloc[:,1:].head()
  # mean0nanmean
  # notetaking = notetaking.replace(0, np.nan)
  group_count = copy_paste['cluster'].value_counts()
  copy_paste = copy_paste.groupby(['cluster']).mean()
  copy_paste['count'] = group_count.sort_index()

  # Execution
  from sklearn.cluster import KMeans
  from collections import Counter
  kmeans = KMeans(n_clusters=5).fit(Execution.iloc[:,1:])
  centroids = kmeans.cluster_centers_
  predict = kmeans.predict(Execution.iloc[:,1:])

  Execution['cluster'] = predict
  # reading1.iloc[:,1:].head()
  # mean0nanmean
  # notetaking = notetaking.replace(0, np.nan)
  group_count = Execution['cluster'].value_counts()
  Execution = Execution.groupby(['cluster']).mean()
  Execution['count'] = group_count.sort_index()


  return notebook_open, code_speed, copy_paste, Execution
```

```python
notebook_open0, code_speed0, copy_paste0, Execution0 = vs_strategy_calculate(df, 0)
notebook_open0 = notebook_open0.sort_values(['count'])
code_speed0 = code_speed0.sort_values(['count'])
copy_paste0 = copy_paste0.sort_values(['count'])
Execution0 = Execution0.sort_values(['count'])
notebook_open1, code_speed1, copy_paste1, Execution1 = vs_strategy_calculate(df, 1)
notebook_open1 = notebook_open1.sort_values(['count'])
code_speed1 = code_speed1.sort_values(['count'])
copy_paste1 = copy_paste1.sort_values(['count'])
Execution1 = Execution1.sort_values(['count'])
```

```python
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import numpy as np
import matplotlib.patches as mpatches

def draw(sequences, strategy_name, behavior, ratio):
  data = sequences
  data = np.array(data)
  unique_chars, matrix = np.unique(data, return_inverse=True)
  color_dict = {'empty': 'blue', 'low': 'powderblue', 'high': 'tomato'}

  # matrix
  plt.figure(figsize=(12,10))
  plt.matshow(matrix.reshape(data.shape), cmap=ListedColormap([color_dict[char] for char in unique_chars]), fignum=1, aspect='auto')

  plt.xticks(np.arange(data.shape[1]), np.arange(data.shape[1]) + 1)
  plt.yticks(np.arange(data.shape[0]), np.arange(data.shape[0]) + 1)
  # plt.xaxis.set_major_formatter(formatter)

  # Set ticks on both sides of axes on
  plt.tick_params(axis="x", bottom=True, top=False, labelbottom=True, labeltop=False)
  plt.yticks([0,1,2,3,4], ratio)

  plt.ylabel(behavior + ' per week', fontsize = 15) # y-axis label with fontsize 15
  # plt.title(strategy_name + ' \n Learning Strategy', loc='left', fontsize = 20)
  plt.rc('xtick', labelsize=40)    # fontsize of the tick labels
  plt.rc('ytick', labelsize=40)    # fontsize of the tick labels
  # custom legend
  reading_patch = mpatches.Patch(color='blue', label = 'nan')
  highlight_patch = mpatches.Patch(color='powderblue', label = 'low')
  notetake_patch = mpatches.Patch(color='tomato', label = 'high')
  # plt.legend(handles=[reading_patch, highlight_patch, notetake_patch], loc='upper left', bbox_to_anchor=(1.04, 1))
  plt.show()
```
